```python
train_data1 = np.array(train_data0.iloc[:]).astype('float32')

test_data1 = np.array(test_data0.iloc[:]).astype('float32')

test_label1 = np.array(test_label0.astype('float32'))
train_label1 = np.array(train_label0.astype('float32'))

train_data1.shape
train_data1.dtype
train_data1.shape[1]
```

```python
#hidden layers including 10,8,6 unit & Output Layer with activation function relu/tanh/drop/regular
```

```python
from keras import models
from keras import layers

def build_model(act):
  model0 = models.Sequential()
  model0.add(layers.Dense(128, activation=act, input_shape=(train_data1.shape[1],)))
  model0.add(layers.Dense(64, activation=act))
  model0.add(layers.Dense(32, activation=act))
  model0.add(layers.Dense(1))
  model0.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
  return model0
```

```python
build_model('relu').summary()
```

```python
build_model('tanh').summary()
```

```python
from keras import  regularizers
def build_model_regular(act):
  model0 = models.Sequential()
  model0.add(layers.Dense(10, activation=act,kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.001),input_shape=(train_data1.shape[1],)))
  model0.add(layers.Dense(8, activation=act,kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.001)))
  model0.add(layers.Dense(6, activation=act,kernel_regularizer=regularizers.l1_l2(l1=0.001, l2=0.001)))
  model0.add(layers.Dense(1))
  model0.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
  return model0
```

```python
build_model_regular('tanh').summary()
```

```python
build_model_regular('relu').summary()
```

```python

from keras import models
from keras import layers
def build_model_drop(act):
  model0 = models.Sequential()
  model0.add(layers.Dense(10, activation=act, input_shape=(train_data1.shape[1],)))
  model0.add(layers.Dropout(0.2))
  model0.add(layers.Dense(8, activation=act))
  model0.add(layers.Dropout(0.2))
  model0.add(layers.Dense(6, activation=act))
  model0.add(layers.Dropout(0.2))
  model0.add(layers.Dense(1))
  model0.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
  return model0
```

```python
build_model_drop('tanh').summary()
```

```python
build_model_drop('relu').summary()
```

```python
#Train the Model with Epochs (100)
```

```python
#KFOLD validation tanh
import numpy as np
k = 4
num_val_samples = len(train_data1) // k
num_epochs = 100
all_scores_tanh = []

for i in range(k):
  print('processing fold #', i)

val_data = train_data1[i * num_val_samples: (i + 1) * num_val_samples]

val_targets = train_label1[i * num_val_samples: (i + 1) * num_val_samples]

partial_train_data = np.concatenate([train_data1[:i * num_val_samples], train_data1[(i + 1) * num_val_samples:]], axis=0)

partial_train_targets = np.concatenate([train_label1[:i * num_val_samples],train_label1[(i + 1) * num_val_samples:]],axis=0)


model = build_model('tanh')
model.fit(partial_train_data, partial_train_targets,epochs=num_epochs, batch_size=1, verbose=0)
val_mse, val_mae = model.evaluate(val_data, val_targets, verbose=0)
all_scores_tanh.append(val_mae)
```
